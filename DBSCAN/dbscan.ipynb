```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn import metrics
from sklearn.datasets import load_iris
x, y = load_iris(return_X_y=True)
```

```python
db = DBSCAN(eps=0.4 , min_samples=3).fit(x)
cl_labels = db.labels_
labels = cl_labels
labels
```

```python
new_x = pd.DataFrame(x)
new_x
```

```python
import matplotlib.pyplot as plt
fig,(ax0,ax1)=plt.subplots(nrows=1,ncols=2,figsize=(20,7),sharey=True);

scatter0 = ax0.scatter( x = new_x[2] ,
            y = new_x[3] ,
            c = y )

ax0.set(title= 'Real Data',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax0.legend(*scatter0.legend_elements(),title='Targets');

#-------------------------------------------------------------

scatter1 = ax1.scatter( x = new_x[2] ,
            y = new_x[3] ,
            c = labels )

ax1.set(title='DBSCAN Clustering',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax1.legend(*scatter1.legend_elements(),title='Cluster Numbers');


fig.suptitle("DBSCAN Clustering Result VS Real Data",fontsize=16,fontweight='bold');
```

```python
labels_true = y

labels = cl_labels
print("Homogeneity: %0.3f" % metrics.homogeneity_score(labels_true, labels))
print("Completeness: %0.3f" % metrics.completeness_score(labels_true, labels))
print("V-measure: %0.3f" % metrics.v_measure_score(labels_true, labels))
print("Adjusted Rand Index: %0.3f"
      % metrics.adjusted_rand_score(labels_true, labels))
print("Adjusted Mutual Information: %0.3f"
      % metrics.adjusted_mutual_info_score(labels_true, labels))
print("Silhouette Coefficient: %0.3f"
      % metrics.silhouette_score(x, labels))
```
