```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import datasets
```

```python
iris = datasets.load_iris()

x = iris.data[:, 2:4]
y = iris.target
```

```python
df = pd.concat([pd.DataFrame(x,columns=['Petal length','petal width']),pd.DataFrame(y,columns=['Species'])],axis=1)

df = df[:56]

new_x = df.drop('Species',axis=1)

new_y = df['Species']
```

```python
from imblearn.over_sampling import SMOTE 

sm = SMOTE(random_state=42)

X_sm, y_sm = sm.fit_resample(new_x, new_y)

print(f'''Shape of X before SMOTE: {new_x.shape}
Shape of X after SMOTE: {X_sm.shape}''')

print('\nBalance of positive and negative classes (%):')
y_sm.value_counts(normalize=True) * 100
```

```python
import matplotlib.pyplot as plt
fig,((ax0, ax1))=plt.subplots(nrows=1,ncols=2,figsize=(19,7),sharey=True );

scatter0 = ax0.scatter(x=new_x['Petal length'],
                       y=new_x['petal width'],
                       c=new_y)

ax0.set(title= 'Before SMOTE Data',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax0.legend(*scatter0.legend_elements(),title='Classses');

#----------------------------------------------------------------

scatter1 = ax1.scatter(x=X_sm['Petal length'],
                       y=X_sm['petal width'],
                       c=y_sm)

ax1.set(title= 'After SMOTE Data',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax1.legend(*scatter1.legend_elements(),title='Classses');
```
