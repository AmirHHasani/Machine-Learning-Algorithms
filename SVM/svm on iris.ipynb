```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm, datasets
import pandas as pd
```

```python
iris = datasets.load_iris()
# Take the first two features. We could avoid this by using a two-dim dataset
X = iris.data[:, 2:4]
y = iris.target
```

```python
X = pd.DataFrame(X)
```

```python
clf1 = svm.SVC(kernel='linear')
clf2 = svm.SVC(kernel='poly',degree = 3)
clf3 = svm.SVC(kernel='rbf')
clf4 = svm.SVC(kernel='sigmoid')
```

```python
a1=clf1.fit(X, y)
a2=clf2.fit(X, y)
a3=clf3.fit(X, y)
a4=clf4.fit(X, y)
```

```python
b1=clf1.predict(X)
b2=clf2.predict(X)
b3=clf3.predict(X)
b4=clf4.predict(X)
```

```python
import matplotlib.pyplot as plt
fig,((ax0, ax1), (ax2, ax3))=plt.subplots(nrows=2,ncols=2,figsize=(20,13),sharey=True , sharex=True);


scatter0 = ax0.scatter( x = X[0] ,
            y = X[1] ,
            c = b1 )

ax0.set(title= 'linear SVM',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax0.legend(*scatter0.legend_elements(),title='Classses');

#-------------------------------------------------------------

scatter1 = ax1.scatter( x = X[0] ,
            y = X[1] ,
            c = b2 )

ax1.set(title='Poly SVM',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax1.legend(*scatter1.legend_elements(),title='Classes');

#-------------------------------------------------------------
scatter2 = ax2.scatter( x = X[0] ,
            y = X[1] ,
            c = b3 )

ax2.set(title='RBF SVM',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax2.legend(*scatter2.legend_elements(),title='Classes');
#--------------------------------------------------------------
scatter3 = ax3.scatter( x = X[0] ,
            y = X[1] ,
            c = b4 )

ax3.set(title='Sigmoid SVM',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax3.legend(*scatter3.legend_elements(),title='Classes');





fig.suptitle("Different SVM Results",fontsize=16,fontweight='bold');
```

```python
fig,ax11 =plt.subplots(figsize=(8,5),sharey=True);


scatter11 = ax11.scatter( x = X[0] ,
            y = X[1] ,
            c = y )

ax11.set(title='Real Data',xlabel = 'petal Length (cm)' ,
       ylabel ='petal Width (cm)' )

ax11.legend(*scatter11.legend_elements(),title='Classes');
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix

target_names = ['Iris-setosa', 'Iris-versicolor' , 'Iris-virginica']

print('Linear',classification_report(y, b1, target_names=target_names))

print('poly',classification_report(y, b2, target_names=target_names))

print('RBF',classification_report(y, b3, target_names=target_names))

print('Sigmoid',classification_report(y, b4, target_names=target_names))
```
